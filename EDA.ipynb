## **data overview**


*   **chroma_tft** : captures pitch class intensity
*   **rmse** : Root Mean Square Energy
*   **spectral_centroid** : indicates where "the center of mass" is
*   **spectral bandwidth** : spread of the spectrum. higher values= more complex timbre
*   **rolloff** : frequency below which a % of total energy is contained
*   **zero_crossing_rate** : how frequently the signal crosses zero
*   **harmony** : harmonic content, often related to tonal components
*   **perceptr_mean** : refers to perceptual tempo or beat-related summary
*   **tempo** : estimated tempo in BPM
*   **mfcc1 - mfcc20** : catch timbral texture and color
*  **genre** : our target colunmn, list of all the genres the track belongs to

## **mapping each cel in the genre to its title to facilitate analysis**

```python
df['genres'] = df['genres'].apply(
    lambda x: [int(i) for i in ast.literal_eval(x)] if isinstance(x, str)
    else [int(i) for i in x] if isinstance(x, list)
    else []
)

id_to_title = dict(zip(genres['genre_id'], genres['title']))

# Map IDs to titles
df['genres_named'] = df['genres'].apply(
    lambda id_list: [id_to_title.get(i) for i in id_list] if isinstance(id_list, list) else []
)
```
